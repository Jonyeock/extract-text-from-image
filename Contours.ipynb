```
import numpy as np
import cv2
from matplotlib import pyplot as plt
```

```
## findContours
## argument1: source image
## argument2: contour retriemal mode
## argument3: contour approximation method

img = cv2.imread('cat.jpg')
imgray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
ret, thresh = cv2.threshold(imgray, 127, 255, 0)
image, contours, hierarchy = cv2.findContours(thresh, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)

# output contours -> Python list of all the contours in the image
# individual contour is a Numpy array of (x, y)
```

```
## draw the contours

# draw all the contours in an image
img = cv2.drawContours(img, contours, -1, (0, 255, 0), 3)

# draw an individual contour
# img = cv2.drawContours(img, contours, 3, (0, 255, 0), 3)
```

```
plt.imshow(img, cmap='gray')
plt.show
```
